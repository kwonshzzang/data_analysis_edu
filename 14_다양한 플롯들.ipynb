```python
import numpy as np
import matplotlib.pyplot as plt

# 그래프를 위한 기본 설정
# 폰트 설정
#plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
# 그래프의 크기
plt.rcParams['figure.figsize'] = 12, 6
# 글자 크기
plt.rcParams['font.size'] = 14
# 폰트 설정시 마이너스 기호가 깨지는 것을 방지한다.
plt.rcParams['axes.unicode_minus'] = False
```

### 선 그래프 그리기
- 어떠한 데이터 흐름에 따라 변화되는 데이터를 표현할때 사용
- 값을 하나만 설정하면 x축은 값의 개수만큼 자동 설정됨
- 값을 두개 설정하면 하나는 x축, 다른 하나는 y축으로 설정됨
- 값의 개수를 동일하게 줘야 함


```python
plt.title('선그래프 1')
# 값 하나를 설정하면 y축으로 적용되고 x축은 값의 개수만큼 자동 설정
plt.plot([2, 3, 1])
plt.plot([1, 2, 3])
plt.show()
```

```python
plt.title('선그래프 2')
# 값을 두 가지로 설정하면 처음 것은 x축, 두번째 것은 y축에 적용됨
plt.plot([1, 2, 3], [2, 3, 1])
plt.plot([1, 2, 3], [1, 2, 3])

# 각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축', rotation=0)

# 각 축의 눈금을 설정함
# 첫번째: 눈금을 표시할 각 축의 위치
# 두번째: 눈금을 표시할 문자열(생략시 숫자로 표시)
plt.xticks([1, 2, 3], ['가', '나', '다'])
plt.yticks([0, 2, 4], ['하', '중', '상'])

plt.show()
```

### 막대그래프
- 데이터의 양을 비교하기 위해 사용함

```python
x = [1, 2, 3]
y = [2, 3, 1]

plt.title('막대 그래프')
plt.bar(x, y)

# 각 축의 이름
plt.xlabel('x 축')
plt.ylabel('y 축')

# 각 축의 눈금 설정
plt.xticks([1, 2, 3], ['가', '나', '다'])
plt.yticks([0, 2, 4], ['하', '중', '상'])

plt.show()
```

### 가로 막대 그래프
- 막대가 많이 나오는 그래프인 경우 좌우로 스크롤해서 봐야 하기 때문에 불편할 수 있음
- 이럴 때는 가로 막대 그래프로 그려줌

```python
x = [1, 2, 3]
y = [2, 3, 1]

plt.title('가로 막대 그래프')
# bar: 첫번째-x축, 두번째-y축
# barh: 첫번째-y축, 두번째-x축
plt.barh(x, y)

# 각 축의 이름
plt.ylabel('x 축')
plt.xlabel('y 축')

# 각 축의 눈금 설정
plt.yticks([1, 2, 3], ['가', '나', '다'])
plt.xticks([0, 2, 4], ['하', '중', '상'])

plt.show()
```

### 파이 차트
- 카테고리별 값의 비율을 비교할때 사용함
- 그래프를 그리기 위한 비율을 자동으로 계산하여 적용해 줌

```python
# 데이터
values = [15, 30, 40, 10]
# 각 파이의 레이블
labels = ['값1', '값2', '값3', '값4']
# 각 파이의 색상
# https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

# 조각이 튀어나오는 정도(0~ 실수값. 1: 파이차트의 반지름 크기)
explode = [0, 0.1, 0, 0]

# 첫번째 값
# labels: 각 파이의 레이블
# colors: 각 파이의 색상
# explode: 각 파이가 튀어나오는 정도
# shadow: True를 설정하면 그림자가 나타남
# startangel: 원의 시작 각도(양수값이면 시계 반대방향, 음수값이면 시계방향)
# counterclock: 시작점부터 그리는 방향(True(기본): 시계 반대방향, False: 시계방향)
# autopct: 파이에서 설정된 비율 값의 양식, 생락하면 비율이 표시되지 않고 설정하면 설정된 양식으로 비율이 나타남 %.자리f%%
plt.pie(values, labels=labels, colors=colors, explode=explode, shadow=True, startangle=90, counterclock=False, autopct='%.2f%%')
plt.show()
```

### 히스토그램
- 막대 그래프와 유사해 보이지만 범위형 데이터를 구간으로 나눠 구간별 양을 비교할 수 있음
- 높이 솟아 있는 부분을 봉우리라고 하고 좌우로 길게 뻗어 있는 부분을 꼬리라고 부름
- 봉우리의 높이와 위치, 꼬리의 길이를 보고 불균형도, 불순도 등등 현재의 데이터 상태를 해석하는 용도로 사용

```python
# -3 ~ 3 사이의 랜덤 숫자를 추출함
# 매개변수로는 개수를 넣어줌
x = np.random.randn(1000)

plt.title('히스토그램')
# bins : 구간의 개수. 기본이 1
plt.hist(x, bins=30)
plt.show()
```

### 산점도
- 2차원 데이터의 상관 관계를 파악할 수 있음
- 분류의 문제에서 이 데이터를 가지고 머신러닝 알고리즘이 분류 예측을 잘 할 수 있는지 확인하는데 사용
- 데이터의 분포 등을 확인하는 용도로 사용
- 만약 중첩되는 구간이 매우 넓으면 머신러닝 알고리즘은 오답을 던질 확률이 높아짐

```python
# 0~1 사이의 랜덤값 100개를 추출함
x1 = np.random.rand(100)
y1 = np.random.rand(100)

x2 = np.random.rand(100)
y2 = np.random.rand(100)

plt.title('산점도')
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
```
