- Matplotlib는 파이썬에서 데이터를 시각화 할 때 사용하는 라이브러리
- 다른 시각화 라이브러리들이 Matplotlib를 기반으로 만들어져 있다.
- 라인플롯, 스케터 플롯 등 다양한 그래프를 그릴 수 있다.
- https://matplotlib.org/stable/gallery/index

```python
import matplotlib.pyplot as plt
```

```python
# 그래프를 위한 기본 설정
# 폰트 설정
#plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
# 그래프의 크기
plt.rcParams['figure.figsize'] = 12, 6
# 글자 크기
plt.rcParams['font.size'] = 14
# 폰트 설정시 마이너스 기호가 깨지는 것을 방지한다.
plt.rcParams['axes.unicode_minus'] = False
```

### 그래프 그리기
- matplotlib는 각 그래프마다 그리는 함수를 제공
- 함수를 사용할때 매개변수에 필요한 정보를 넣어주면 그 정보를 기반으로 그래프를 그려줌
- 그려지는 그래프의 종류는 그에 맞는 함수를 호출함으로써 결정

```python
# plot 함수를 이용해 선 그래프를 그림
plt.plot([0, 1, 2, 3], [1, 4, 9, 16])
plt.show()
```

```python
# bar 함수를 이용해 막대 그래프를 그림
plt.bar([0, 1, 2, 3], [1, 4, 9, 16])
plt.show()
```

### 다수의 그래프 그리기
- plt.show() 함수를 호출하거나 셀이 끝나면 그때까지 그린 모든 그래프는 하나의 좌표평면에 표현
- subplot을 사용하면 다수의 좌표평면을 그리드 형태로 그릴 수 있음

```python
# 그래프를 그리는 함수를 여러개 호출하고 plt.show()를 호출함
plt.plot([0, 1, 2, 3],[1, 4, 9, 16])
plt.plot([0, 1, 2, 3], [16, 9, 4, 1])
plt.bar([0, 1, 2, 3], [1, 4, 9, 16])
plt.bar([0, 1, 2, 3], [16, 9, 4, 1])
plt.show()
```

```python
# 중간 중간에 plt.show()를 호출
plt.plot([0, 1, 2, 3],[1, 4, 9, 16])
plt.show()

plt.plot([0, 1, 2, 3], [16, 9, 4, 1])
plt.show()

plt.bar([0, 1, 2, 3], [1, 4, 9, 16])
plt.show()

plt.bar([0, 1, 2, 3], [16, 9, 4, 1])
plt.show()
```

```python
# subplot: 다수의 그래프를 모아 한번에 그린다.
# 전체 행의 수와 열의 수를 정해주고 몇 번째 그릴 것인지 정해줌
# 가로에서 우측으로 위에서 아래로 순서가 정해지며 1부터 1씩 증가되는 값을 설정
# (전체 행의 수, 전체 열의 수, 몇번째 인가..)
plt.subplot(2, 2, 1)
plt.plot([0, 1, 2, 3], [1, 4, 9, 16])

plt.subplot(2, 2, 2)
plt.plot([0, 1, 2, 3], [16, 9, 4, 1])

plt.subplot(2, 2, 3)
plt.bar([0, 1, 2, 3], [1, 4, 9, 16])

plt.subplot(2, 2, 4)
plt.bar([0, 1, 2, 3], [16, 9, 4, 1])
plt.show()
```

### 그래프 제목

```python
# 그래프의 타이틀을 설정
plt.title('그래프의 제목')
plt.plot([0, 1, 2, 3], [1, 4, 9, 16])
plt.show()
```

```python
# 타이틀 옵션
d1 = {
    'color':'red'
}

# fontdict : 표시되는 문자열에 대한 설정
# loc: 타이틀의 위치, center(기본), left, right
# y: 타이틀의 y좌표, 1(기본) - 제일 위, 0 - 제일아래
# x: 타이틀의 x좌표, 1은 좌표평면의 가로 길이 크기를 의미
plt.title('그래프의 제목', fontdict=d1, loc='left', y = 0.5, x = -0.3)
plt.plot([0, 1, 2, 3], [1, 4, 9, 16])
plt.show()
```

```python
# plt.title: 각 서브 플롯의 타이틀을 설정
# plt.suptitle: 전체의 타이틀을 설정

plt.subplot(2, 2, 1)
plt.title('sub 1')
plt.plot([0, 1, 2, 3], [1, 4, 9, 16])

plt.subplot(2, 2, 2)
plt.title('sub 2')
plt.plot([0, 1, 2, 3], [16, 9, 4, 1])

plt.subplot(2, 2, 3)
plt.title('sub 3')
plt.bar([0, 1, 2, 3], [1, 4, 9, 16])

plt.subplot(2, 2, 4)
plt.title('sub 4')
plt.bar([0, 1, 2, 3], [16, 9, 4, 1])

plt.suptitle('전체')

plt.show()
```

### 각 축의 이름을 설정

```python
plt.plot([0, 1, 2, 3], [1, 4, 9, 16])

# rotation: 회전

# x축의 이름
plt.xlabel('x 축')
# y축의 이름
plt.ylabel('y 축', rotation=0)

plt.show()
```

### 폰트 설정
- plt.toParams['font_family']로 설정하면 전체에 대한 기본 설정을 할 수 있음
- 폰트를 직접 지정하여 각각 다르게 표시할 수 있음
- os의 시스템 폰트를 사용하는 방법과 폰트를 직접 지정하여 사용하는 방법이 있음

```python
import matplotlib.font_manager as fm

plt.plot([0, 1, 2, 3], [1, 4, 9, 16])
# 폰트의 경로는 /Users/kwonsoonhyun/Library/fonts 플더에서 폰트를 찾아
# 마우스 우클릭 속성에서 경로를 확인해야 함.
path1 = '/Users/kwonsoonhyun/Library/fonts/D2Coding-Ver1.3.2-20180524-all.ttc'
# 폰트의 이름을 추출
prop1 = fm.FontProperties(fname=path1)
name1 = prop1.get_name()

print(name1)

option1 = {
    'family': name1,
    'size': 24,
    'color': 'blue'
}

plt.title('시각화', fontdict=option1)
plt.xlabel('x 축', fontdict=option1)
plt.ylabel('y 축', rotation=0)

plt.show()
```
